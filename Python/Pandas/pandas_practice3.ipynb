# Group By in pandas

```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data.csv")
df
```

```python
gb = df.groupby("Weather")
gb
```

```python
# get all the groups

for weather, weather_df in gb:
    print(weather)
    print()
    print(weather_df)
    print()
    print()
```

```python
# to get a specific group

gb.get_group("Clear")
```

```python
# get max in diff groups

gb.max()
```

```python
"""
The above process is called Split-Apply-Combine

split using grouping by
Appling some functionality to those groups lke avg, sum etc
Combine the data together after applyting
"""

# Split
gb = df.groupby("Weather")

# Apply and Combine
gb.max()
```
