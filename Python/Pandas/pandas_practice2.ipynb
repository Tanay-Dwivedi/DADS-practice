# Handle missing data in pandas

```python
import pandas as pd
```

```python
df = pd.read_csv("weather_data.csv")
df
```

```python
"""
to convert the data/time column from string to date and time
"""

df = pd.read_csv("weather_data.csv", parse_dates = ["Date/Time"])
df
```

```python
df.set_index("Date/Time", inplace = True)
```

```python
df
```

```python
# fillna(value) method

dff = df.fillna(0)
dff
```

```python
"""
if want to fill diff columns having no data with diff values
"""

dff = df.fillna({
    "column_name": value,
    .
    .
    . # other column names with the calues to be replaced with
})
dff
```

```python
"""
to carry forward the value to be filled to the next row of the same column
means - same upar wali value next row pe v copy
"""

dff = df.ffill()
dff
```

```python
"""
to carry backward the value to be filled from the previous row row of the same column
means - same nenche wali value prev cell me copy
"""

dff = df.bfill()
dff
```

```python
"""
previous col ki value copy to next column too
"""

dff = df.ffill(axis = "columns")
dff
```

```python
"""
next col ki value copy to prev column too
"""

dff = df.bfill(axis = "columns")
dff
```

```python
"""
limit use only when we have to use ffill or bfill only limited number of times
"""

dff = df.ffill(axis = "columns", limit = 2)
dff
```

```python
"""
if we want the values to be between the next and prev values of the cell, we use
infer_objects(copy=False) function
can only infer dtype datas
"""

# dff = df.interpolate() -- depricated
dff = df.infer_objects(copy=False)
dff
```

```python
# to drop the rows with NA values

dff = df.dropna()
dff
```

```python
"""
conditional dropping of rows
"""

dff = df.dropna(how="all") # drop rows with all na values
dff
```

```python
"""
drop rows with more than 1 na values
thresh = 1
"""

dff = df.dropna(thresh = 1)
dff
```

```python
"""
if certain dates are not there and we want to add it to the file
given it should be only date in the column
"""

dt = pd.date_range("2012-01-01", "2012-01-10")
idx = pd.DataFrame(dt)
dff = df.reindex(idx)
dff
```
