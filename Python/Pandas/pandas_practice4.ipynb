# Pivot

```python
import pandas as pd
```

```python
df = pd.DataFrame([
    {'date': '5/1/2017', 'city': 'new york', 'temperature': '65', 'humidity': '56'},
    {'date': '5/2/2017', 'city': 'new york', 'temperature': '66', 'humidity': '58'},
    {'date': '5/3/2017', 'city': 'new york', 'temperature': '68', 'humidity': '60'},
    {'date': '5/1/2017', 'city': 'mumbai', 'temperature': '75', 'humidity': '80'},
    {'date': '5/2/2017', 'city': 'mumbai', 'temperature': '78', 'humidity': '83'},
    {'date': '5/3/2017', 'city': 'mumbai', 'temperature': '82', 'humidity': '85'},
    {'date': '5/1/2017', 'city': 'beijing', 'temperature': '80', 'humidity': '26'},
    {'date': '5/2/2017', 'city': 'beijing', 'temperature': '77', 'humidity': '30'},
    {'date': '5/3/2017', 'city': 'beijing', 'temperature': '79', 'humidity': '35'}
])

df
```

```python
df.pivot(index="date", columns="city")
```

```python
# if want only humidity in the table

df.pivot(index="date", columns="city", values = "humidity")
```

# Pivot Table

```python
dff = pd.DataFrame([
    {'date': '5/1/2017', 'city': 'new york', 'temperature': 65, 'humidity': 56},
    {'date': '5/1/2017', 'city': 'new york', 'temperature': 61, 'humidity': 54},
    {'date': '5/2/2017', 'city': 'new york', 'temperature': 70, 'humidity': 60},
    {'date': '5/2/2017', 'city': 'new york', 'temperature': 72, 'humidity': 62},
    {'date': '5/1/2017', 'city': 'mumbai', 'temperature': 75, 'humidity': 80},
    {'date': '5/1/2017', 'city': 'mumbai', 'temperature': 78, 'humidity': 83},
    {'date': '5/2/2017', 'city': 'mumbai', 'temperature': 82, 'humidity': 85},
    {'date': '5/2/2017', 'city': 'mumbai', 'temperature': 80, 'humidity': 26}
])
dff
```

```python
dff.pivot_table(index="city",columns="date", aggfunc = "count")
```

```python
# it is by default

dff.pivot_table(index="city",columns="date", aggfunc = "mean")
```

```python
# to provide aggregate statistocs of all

dff.pivot_table(index="city",columns="date", margins = True)
```

```python
df = pd.DataFrame([
    {'date': '5/1/2017', 'city': 'new york', 'temperature': 65, 'humidity': 56},
    {'date': '5/2/2017', 'city': 'new york', 'temperature': 61, 'humidity': 54},
    {'date': '5/3/2017', 'city': 'new york', 'temperature': 70, 'humidity': 60},
    {'date': '12/1/2017', 'city': 'new york', 'temperature': 30, 'humidity': 50},
    {'date': '12/2/2017', 'city': 'new york', 'temperature': 28, 'humidity': 52},
    {'date': '12/3/2017', 'city': 'new york', 'temperature': 25, 'humidity': 51}
])
df
```

```python
df["date"] = pd.to_datetime(df["date"])
df
```

```python
type(df["date"][0])
```

```python
df.pivot_table(index = pd.Grouper(freq = "M", key = "date"), columns = "city")
```

# Melt in pandas

```python
df = pd.DataFrame({
    'day': ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday'],
    'chicago': [32, 30, 28, 22, 30, 20, 25],
    'chennai': [75, 77, 75, 82, 83, 81, 77],
    'berlin': [41, 43, 45, 38, 30, 45, 47]
})
df
```

```python
dff = pd.melt(df, id_vars=["day"])
dff
```

```python
dff[dff["variable"]=="berlin"]
```

```python
dff = pd.melt(df, id_vars=["day"], var_name = "city", value_name = "temperature")
dff
```

# Crosstable

```python
import pandas as pd
```

```python
df = pd.read_csv("survey.csv")
df
```

```python
pd.crosstab(df.Nationality, df.Handedness)
```

```python
pd.crosstab(df.Sex, df.Handedness)
```

```python
pd.crosstab(df.Sex, df.Handedness, margins = True)
```

```python
pd.crosstab(df.Sex, [df.Handedness,df.Nationality], margins = True)
```

```python
pd.crosstab([df.Sex,df.Nationality], df.Handedness, margins = True)
```

```python
pd.crosstab(df.Sex, df.Handedness, margins = True)
```
