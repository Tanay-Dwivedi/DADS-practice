# Step Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y)
plt.show()
```

```python
# line change

x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y, linefmt  =":")
plt.show()
```

```python
# marker change

x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y, linefmt  =":", markerfmt = "r*")
plt.show()
```

```python
# change base line

x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y, linefmt  =":", markerfmt = "r*", bottom = 3)
plt.show()
```

```python
# base line color change

x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y, linefmt  =":", markerfmt = "r*", bottom = 3, basefmt = "g")
plt.show()
```

```python
# labels

x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y, linefmt  =":", markerfmt = "r*", bottom = 3,
        label = "some stats")
plt.legend()
plt.show()
```

```python
# orientation

x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y, linefmt  =":", markerfmt = "r*", bottom = 3,
        label = "some stats", orientation = "horizontal")
plt.legend()
plt.show()
```
