# Step Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y)
plt.show()
```

```python
# line change

x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y, linefmt  =":")
plt.show()
```

```python
# marker change

x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y, linefmt  =":", markerfmt = "r*")
plt.show()
```

```python
# change base line

x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y, linefmt  =":", markerfmt = "r*", bottom = 3)
plt.show()
```

```python
# base line color change

x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y, linefmt  =":", markerfmt = "r*", bottom = 3, basefmt = "g")
plt.show()
```

```python
# labels

x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y, linefmt  =":", markerfmt = "r*", bottom = 3,
        label = "some stats")
plt.legend()
plt.show()
```

```python
# orientation

x = [1,2,3,4,5,6]
y = [2,5,3,1,7,3]

plt.stem(x,y, linefmt  =":", markerfmt = "r*", bottom = 3,
        label = "some stats", orientation = "horizontal")
plt.legend()
plt.show()
```

# Box or Whisker plot

```python
import matplotlib.pyplot as plt
```

```python
x = [10,20,30,40,50,60,70]
plt.boxplot(x)
plt.show()
```

```python
# change box shape

x = [10,20,30,40,50,60,70]
plt.boxplot(x, notch = True)
plt.show()
```

```python
# orientation

x = [10,20,30,40,50,60,70]
plt.boxplot(x, vert = False)
plt.show()
```

```python
# box width

x = [10,20,30,40,50,60,70]
plt.boxplot(x, widths = 0.25)
plt.show()
```

```python
# label

x = [10,20,30,40,50,60,70]
plt.boxplot(x, labels = ["Java"])
plt.show()
```

```python
# bos fill

x = [10,20,30,40,50,60,70]
plt.boxplot(x, labels = ["Java"], patch_artist = True)
plt.show()
```

```python
# show mean

x = [10,20,30,40,50,60,70]
plt.boxplot(x, labels = ["Java"], showmeans = True)
plt.show()
```

```python
# outliers

x = [10,20,30,40,50,60,70, 120]
plt.boxplot(x, labels = ["Java"])
plt.show()
```

```python
# wisk join with outlier

x = [10,20,30,40,50,60,70,120]
plt.boxplot(x, labels = ["Java"], whis = 2, showmeans = True)
plt.show()
```

```python
# change colors

x = [10,20,30,40,50,60,70,120]
plt.boxplot(x, labels = ["Java"], showmeans = True, sym = "r*")
plt.show()
```

```python
# multiple bos plots

x = [10,20,30,40,50,60,70,120]
y = [23,23,45,22,34,56,44,23]
plt.boxplot(x, labels = ["Java"], showmeans = True, sym = "r*")
plt.boxplot(y, labels = ["Java"], showmeans = True, sym = "r*")
plt.show()
```

```python
x = [10,20,30,40,50,60,70,120]
y = [10,20,30,40,50,30,60,90]
z = [x,y]
plt.boxplot(z, labels = ["Java", "Python"], showmeans = True, sym = "r*")
plt.show()
```

# Area vs Stack Plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
a1 = [2,3,1,5,2]
a2 = [4,1,2,5,2]
a3 = [2,4,2,3,1]
a4 = [1,5,3,1,2]

plt.stackplot(x,a1,a2,a3,a4)
plt.show()
```

```python
# labels

x = [1,2,3,4,5]
a1 = [2,3,1,5,2]
a2 = [4,1,2,5,2]
a3 = [2,4,2,3,1]
a4 = [1,5,3,1,2]

ls = ["C", "C++", "Java", "Python"]
plt.stackplot(x,a1,a2,a3,a4, labels = ls)
plt.legend()
plt.show()
```

```python
# change baseline

x = [1,2,3,4,5]
a1 = [2,3,1,5,2]
a2 = [4,1,2,5,2]
a3 = [2,4,2,3,1]
a4 = [1,5,3,1,2]

ls = ["C", "C++", "Java", "Python"]
plt.stackplot(x,a1,a2,a3,a4, labels = ls, baseline = "sym")
plt.legend()
plt.show()
```

```python
# change baseline

x = [1,2,3,4,5]
a1 = [2,3,1,5,2]
a2 = [4,1,2,5,2]
a3 = [2,4,2,3,1]
a4 = [1,5,3,1,2]

ls = ["C", "C++", "Java", "Python"]
plt.stackplot(x,a1,a2,a3,a4, labels = ls, baseline = "wiggle")
plt.legend()
plt.show()
```

```python
# title

x = [1,2,3,4,5]
a1 = [2,3,1,5,2]
a2 = [4,1,2,5,2]
a3 = [2,4,2,3,1]
a4 = [1,5,3,1,2]

ls = ["C", "C++", "Java", "Python"]
plt.stackplot(x,a1,a2,a3,a4, labels = ls, baseline = "sym")
plt.title("Stats")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
# grid adding

x = [1,2,3,4,5]
a1 = [2,3,1,5,2]
a2 = [4,1,2,5,2]
a3 = [2,4,2,3,1]
a4 = [1,5,3,1,2]

ls = ["C", "C++", "Java", "Python"]
plt.stackplot(x,a1,a2,a3,a4, labels = ls, baseline = "sym")
plt.title("Stats")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.grid()
plt.legend()
plt.show()
```

# Fill_between plot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.title("plots")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
x = [1,2,3,4,5]
y = [1,2,3,4,5]

plt.plot(x,y)
plt.fill_between(x=[2,4],y1=2, y2=4)
plt.title("plots")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

```python
# where
import numpy as np

x = np.array([1,2,3,4,5])
y = np.array([1,2,3,4,5])

plt.plot(x,y)
plt.fill_between(x,y, where = (x>=2) & (x<=4), alpha = 0.4)
plt.title("plots")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.show()
```

# Subplot

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y)
plt.pie([1])
plt.show()
```

```python
# caution the graphs are merging hence we use subplot

x = [1,2,3,4]
y = [1,2,3,4]

plt.plot(x,y)
x1 = [10,20,30,40]
plt.pie(x1)
plt.show()
```

```python
x = [1,2,3,4]
y = [1,2,3,4]
plt.subplot(1,2,1)
plt.plot(x,y)
x1 = [10,20,30,40]
plt.subplot(1,2,2)
plt.pie(x1)
plt.show()
```

```python
x = [1,2,3,4]
y = [1,2,3,4]
plt.subplot(2,2,1)
plt.plot(x,y)


x1 = [10,20,30,40]
plt.subplot(2,2,2)
plt.pie(x1)


plt.subplot(2,2,3)
plt.bar(x,y)


plt.subplot(2,2,4)
plt.scatter(x,y)


plt.show()
```

# Savefig

> To save the graphs

```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [2,4,5,1,3]

plt.plot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5]
y = [2,4,5,1,3]

plt.plot(x,y)
plt.savefig("line_plot")
plt.show()
```

```python
# saved fig modifications

x = [1,2,3,4,5]
y = [2,4,5,1,3]

plt.plot(x,y)
plt.savefig("line_plot", dpi = 1000)
plt.show()
```

```python
# saved fig modifications

x = [1,2,3,4,5]
y = [2,4,5,1,3]

plt.plot(x,y)
plt.savefig("line_plot.pdf", dpi = 1000, transparent = True)
plt.show()
```
