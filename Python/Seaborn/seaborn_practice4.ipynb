# Pair Plot

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
df = sb.load_dataset("tips")
```

```python
df.head()
```

```python
sb.pairplot(df)
plt.show()
```

```python
# add hue

sb.pairplot(df, hue = "sex")
plt.show()
```

```python
sb.pairplot(df,vars = ["total_bill", "tip"], hue = "sex")
plt.show()
```

```python
# set order

sb.pairplot(df, hue = "sex", hue_order = ["Female", "Male"])
plt.show()
```

```python
# change color palette

sb.pairplot(df, hue = "sex", palette = "Accent")
plt.show()
```

```python
# x axis graphs

sb.pairplot(df, hue = "sex", x_vars = ["total_bill", "tip"])
plt.show()
```

```python
# y axis graphs

sb.pairplot(df, hue = "sex", y_vars = ["total_bill", "tip"])
plt.show()
```

```python
# change graph kind

sb.pairplot(df, hue = "sex", hue_order = ["Female", "Male"], kind = "reg")
plt.show()
```

```python
# change graph kind

sb.pairplot(df, hue = "sex", hue_order = ["Female", "Male"], kind = "kde")
plt.show()
```

```python
# make uniform similar graph kinds

sb.pairplot(df, hue = "sex", hue_order = ["Female", "Male"], kind = "kde",
           diag_kind = "kde")
plt.show()
```

```python
# change markers

sb.pairplot(df, hue = "sex", hue_order = ["Female", "Male"],
           markers = ["*", "<"])
plt.show()
```

# Strip Plot

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
df = sb.load_dataset("tips")
```

```python
df.head()
```

```python
sb.stripplot(x = "day", y = "total_bill", data = df)
plt.show()
```

```python
# add hue

sb.stripplot(x = "day", y = "total_bill", data = df, hue = "sex")
plt.show()
```

```python
# change color palette

sb.stripplot(x = "day", y = "total_bill", data = df, hue = "sex",
            palette = "Accent")
plt.show()
```

```python
# adding line width

sb.stripplot(x = "day", y = "total_bill", data = df, hue = "sex",
             linewidth = 0.5)
plt.show()
```

```python
# edge color

sb.stripplot(x = "day", y = "total_bill", data = df, hue = "sex",
            edgecolor = "r")
plt.show()
```

```python
# scatter the points

sb.stripplot(x = "day", y = "total_bill", data = df, hue = "sex",
            jitter = 5)
plt.show()
```

```python
# size

sb.stripplot(x = "day", y = "total_bill", data = df, hue = "sex",
            size = 10)
plt.show()
```

```python
# adding marker

sb.stripplot(x = "day", y = "total_bill", data = df, hue = "sex",
            marker = "*")
plt.show()
```

```python
# change opacity

sb.stripplot(x = "day", y = "total_bill", data = df, hue = "sex", alpha = 0.5)
plt.show()
```

```python
# single plot graph

sb.stripplot(x = df["total_bill"])
plt.show()
```

# Box Plot

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
df = sb.load_dataset("tips")
```

```python
df.head()
```

```python
sb.set(style = "whitegrid")
sb.boxplot(x = "day", y = "total_bill", data = df)
plt.show()
```

```python
# add hue

sb.boxplot(x = "day", y = "total_bill", data = df, hue = "time")
plt.show()
```

```python
# change color palette

sb.boxplot(x = "day", y = "total_bill", data = df, hue = "time",
          palette='dark:g')
plt.show()
```

```python
# change order

sb.boxplot(x = "day", y = "total_bill", data = df, hue = "time",
          palette='dark:g', order = ["Fri", "Sun", "Thur"])
plt.show()
```

```python
# add mean

sb.boxplot(x = "day", y = "total_bill", data = df, hue = "time",
          palette='dark:r', showmeans = True)
plt.show()
```

```python
# add line width

sb.boxplot(x = "day", y = "total_bill", data = df, hue = "time",
          palette='dark:g', linewidth = 3)
plt.show()
```

```python
# change orientation

sb.boxplot(y = "day", x = "total_bill", data = df, hue = "time",
          palette='dark:g')
plt.show()
```

```python
# change orientation using orient

sb.boxplot(data = df, palette="Accent", orient = "h")
plt.show()
```

```python
# single box plot

sb.boxplot(x = df["total_bill"])
plt.show()
```

```python
# single box plot

sb.boxplot(y = df["total_bill"])
plt.show()
```
