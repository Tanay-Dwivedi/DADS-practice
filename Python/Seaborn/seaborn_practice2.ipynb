# Histogram Plot

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
df = sb.load_dataset("penguins")
```

```python
df.head()
```

```python
sb.displot(df["bill_length_mm"])
plt.show()
```

```python
# adding bins

sb.displot(df["bill_length_mm"], bins=[30,40,50,60])
plt.show()
```

```python
# kernel density estimate

sb.displot(df["bill_length_mm"], kde = True)
plt.show()
```

```python
# add kde density lines

sb.displot(df["bill_length_mm"], kde = True, rug = True)
plt.show()
```

```python
# change color 

sb.displot(df["bill_length_mm"], kde = True, color = "b")
plt.show()
```

```python
# change the x axis to log scale

sb.displot(df["bill_length_mm"], kde = True, color = "b", log_scale = True)
plt.show()
```

# Scatter Plot

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
df = sb.load_dataset("penguins").head(25)
```

```python
df.head()
```

```python
sb.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = df)
plt.show()
```

```python
# differentiate the data

sb.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = df,
              hue = "sex")
plt.show()
```

```python
# change styli

sb.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = df,
              style = "sex")
plt.show()
```

```python
# change size

sb.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = df,
              style = "sex", size = "sex")
plt.show()
```

```python
# give particular size

sb.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = df,
              style = "sex", size = "sex", sizes=(100,40))
plt.show()
```

```python
# change color

sb.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = df,
              hue = "sex", palette = "Accent")
plt.show()
```

```python
# change opacity

sb.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = df,
              hue = "sex", alpha = 0.5)
plt.show()
```

```python
# adding markers

mar = {"Male":"*", "Female":">"}
sb.scatterplot(x = "bill_length_mm", y = "bill_depth_mm", data = df,
              hue = "sex", palette = "Accent", markers = mar)
plt.show()
```

# Heatmap Plot

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
import numpy as np
```

```python
df = sb.load_dataset("anagrams")
```

```python
df.head()
```

```python
ds = np.linspace(1,20,50).reshape(5,10)
ds
```

```python
sb.heatmap(ds)
plt.show()
```

```python
df.tail()
```

```python
nd = df.drop(columns = "attnr", axis = 1).head(10)
nd
```

```python
sb.heatmap(nd)
plt.show()
```

```python
# min and max

sb.heatmap(nd, vmin = 0, vmax = 10)
plt.show()
```

```python
# change color palette

sb.heatmap(nd, vmin = 0, vmax = 10, cmap = "Accent")
plt.show()
```

```python
# show values

sb.heatmap(nd, vmin = 0, vmax = 10, annot = True)
plt.show()
```

```python
nar = np.linspace(1,10,10).reshape(2,5)
nar
```

```python
sb.heatmap(nar,  vmin = 0, vmax = 10, annot = True)
plt.show()
```

```python
# strig type annotations

sar = np.array([["a1","a2","a3","a4","a5"],
                ["b1","b2","b3","b4","b5"]])
sar
```

```python
sb.heatmap(nar,  vmin = 0, vmax = 10, annot = sar, fmt = "s")
plt.show()
```

```python
# change annotations style

sty = {"fontsize":10, "color":"g"}
sb.heatmap(nar,  vmin = 0, vmax = 10, annot = True, annot_kws = sty)
plt.show()
```

```python
# heatmap box distance

sty = {"fontsize":10, "color":"g"}
sb.heatmap(nar,  vmin = 0, vmax = 10, annot = True, annot_kws = sty,
           linewidth = 5, linecolor = "b")
plt.show()
```

```python
# remove color bar

sty = {"fontsize":10, "color":"g"}
sb.heatmap(nar,  vmin = 0, vmax = 10, annot = True, annot_kws = sty,
          cbar = False)
plt.show()
```

```python
# add labels and title

sty = {"fontsize":10, "color":"g"}
v = sb.heatmap(nar,  vmin = 0, vmax = 10, annot = True, annot_kws = sty)
v.set(xlabel = "x-axis", ylabel = "y-axis", title = "title")
sb.set(font_scale = 10)
plt.show()
```
