# Line Plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
var = [1,2,3,4,5,6,7]
var1 = [3,4,2,5,6,1,7]
```

```python
# using matplotlib

plt.plot(var, var1)
plt.show()
```

```python
# using seaborn

import pandas as pd

df = pd.DataFrame({"var":var, "var1":var1})
sns.lineplot(x = "var", y = "var1", data = df)
plt.show()
```

```python
# using seaborn {same but with no x and y labels}

import pandas as pd

sns.lineplot(x = var, y = var1)
plt.show()
```

```python
# using datasets

ds = seaborn.load_dataset("penguins").head(25)
```

```python
ds.head()
```

```python
sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = ds)
plt.show()
```

```python
# adding hue to the lines

sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = ds, hue = "sex")
plt.show()
```

```python
# changing size of the plot

sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = ds, hue = "sex",
            size = 10)
plt.show()
```

```python
# changing the line style

sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = ds, hue = "sex",
             style = "sex")
plt.show()
```

```python
# changing the plot line colors

sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = ds, hue = "sex",
             style = "sex", palette = "Accent")
plt.show()
```

```python
# adding the marker points

sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = ds, hue = "sex",
             style = "sex", palette = "Accent", markers = ["o","*"])
plt.show()
```

```python
# removing the dashed line

sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = ds, hue = "sex",
             style = "sex", palette = "Accent", markers = ["o","*"],
            dashes = False)
plt.show()
```

```python
# changing the legends

sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = ds, hue = "sex",
             style = "sex", palette = "Accent", markers = ["o","*"],
            dashes = False, legend = False)
plt.show()
```

```python
# adding grid

sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = ds, hue = "sex",
             style = "sex", palette = "Accent", markers = ["o","*"])
plt.grid()
plt.show()
```

```python
# adding title and labels

# adding grid

sns.lineplot(x = "bill_length_mm", y = "bill_depth_mm", data = ds, hue = "sex",
             style = "sex", palette = "Accent", markers = ["o","*"])
plt.title("Seaborn Line Plot")
plt.xlabel("bill length")
plt.ylabel("bill depth")
plt.show()
```

# Bar Plot

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = seaborn.load_dataset("penguins")
```

```python
df.head()
```

```python
sns.barplot(x = df["island"], y = df["bill_length_mm"])
plt.show()
```

```python
sns.barplot(x = df.island, y = df.bill_length_mm)
plt.show()
```

```python
sns.barplot(x = "island", y = "bill_length_mm", data = df)
plt.show()
```

```python
# adding hue

sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex")
plt.show()
```

```python
# order of the graph

order = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex",
           order = order)
plt.show()
```

```python
# order of legends

order = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex",
           order = order, hue_order = ["Female", "Male"])
plt.show()
```

```python
# confidence interval size
# ci is depricated instead use errorbar

order = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex",
           order = order, hue_order = ["Female", "Male"],
           errorbar=('ci', 25))
plt.show()
```

```python
# n_boot same as ci

order = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex",
           order = order, hue_order = ["Female", "Male"],
            n_boot = 2)
plt.show()
```

```python
# changing orientation

sns.barplot(x = "bill_depth_mm", y = "bill_length_mm", data = df,
            orient = "horizontal")
plt.show()
```

```python
# changing colors

sns.barplot(x = "island", y = "bill_length_mm", data = df,
           color = "b")
plt.show()
```

```python
# using palette to change colors

order = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex",
           order = order, hue_order = ["Female", "Male"],
           palette = "Accent")
plt.show()
```

```python
# adding saturation

order = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex",
           order = order, hue_order = ["Female", "Male"],
           palette = "Accent", saturation = 0.3)
plt.show()
```

```python
# change color of ci
# errcolor is depricated instead use err_kws

order = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex",
           order = order, hue_order = ["Female", "Male"],
           palette = "Accent", err_kws={'color': 'y'})
plt.show()
```

```python
# changing capsize

order = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex",
           order = order, hue_order = ["Female", "Male"],
           palette = "Accent", err_kws={'color': 'y'}, capsize = 0.2)
plt.show()
```

```python
# changing the opacity

order = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex",
           order = order, hue_order = ["Female", "Male"],
           palette = "Accent", err_kws={'color': 'y'}, alpha = 0.5)
plt.show()
```

```python
# adding grid

sns.set(style = "darkgrid")
order = ["Dream", "Torgersen", "Biscoe"]
sns.barplot(x = "island", y = "bill_length_mm", data = df, hue = "sex",
           order = order, hue_order = ["Female", "Male"],
           palette = "Accent", err_kws={'color': 'y'})
plt.show()
```
