# Count Plot

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
df = sb.load_dataset("tips")
```

```python
df.head()
```

```python
sb.countplot(x = "sex", data = df)
plt.show()
```

```python
sb.barplot(x = "sex", y = "size", data = df)
plt.show()
```

```python
# adding hue

sb.countplot(x = "sex", data = df, hue = "smoker")
plt.show()
```

```python
# change orientation

sb.countplot(y = "sex", data = df, hue = "smoker")
plt.show()
```

```python
# change color

sb.countplot(x = "sex", data = df, hue = "smoker", palette = "Accent")
plt.show()
```

```python
# give color
# color is depricated, use palette only

sb.countplot(x = "sex", data = df, hue = "smoker", palette="dark:r")
plt.show()
```

```python
# adding saturation and alpha

sb.countplot(x = "sex", data = df, hue = "smoker", palette="dark:r",
             saturation = 0.5, alpha = 0.5)
plt.show()
```

# Violin Plot

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
df = sb.load_dataset("tips")
```

```python
df.head()
```

```python
sb.violinplot(x = "day", y = "total_bill", data = df)
plt.show()
```

```python
# adding hue

sb.violinplot(x = "day", y = "total_bill", data = df, hue = "time")
plt.show()
```

```python
# adding line width

sb.violinplot(x = "day", y = "total_bill", data = df, hue = "time",
             linewidth = 1.1)
plt.show()
```

```python
# adding color palette

sb.violinplot(x = "day", y = "total_bill", data = df, hue = "time",
             palette = "Accent")
plt.show()
```

```python
# setting the order

order = ["Dinner", "Lunch"]
sb.violinplot(x = "time", y = "total_bill", data = df, order = order)
plt.show()
```

```python
# add saturation

sb.violinplot(x = "day", y = "total_bill", data = df, hue = "time",
             saturation = 0.4)
plt.show()
```

```python
# add color
# color is depricated use palette instead

sb.violinplot(x = "day", y = "total_bill", data = df, hue = "time",
             palette='dark:r')
plt.show()
```

```python
# split data

sb.violinplot(x = "day", y = "total_bill", data = df, hue = "sex", split = True)
plt.show()
```

```python
# scale param
# scale is depricated instead use density_norm

sb.violinplot(x = "day", y = "total_bill", data = df, hue = "sex",
              split = True, density_norm='count')
plt.show()
```

```python
# change orientation

sb.violinplot(y = "day", x = "total_bill", data = df, hue = "sex",
              split = True, density_norm='count')
plt.show()
```

```python
# inner pram

sb.violinplot(x = "day", y = "total_bill", data = df, hue = "sex",
              split = True, density_norm='count', inner = "quart")
plt.show()
```

```python
# single violin plot

sb.violinplot(x = df["total_bill"])
plt.show()
```

```python
# single violin plot

sb.violinplot(y = df["total_bill"])
plt.show()
```
