# Cat Plot

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
df = sb.load_dataset("tips")
```

```python
df.head()
```

```python
sb.catplot(x = "size", y = "tip", data = df)
plt.show()
```

```python
# add hue

sb.catplot(x = "size", y = "tip", data = df, hue = "sex")
plt.show()
```

```python
# kind graph

sb.catplot(x = "size", y = "tip", data = df, hue = "sex",
          kind = "bar")
plt.show()
```

```python
sb.catplot(x = "tip", y = "size", data = df, hue = "sex")
plt.show()
```

```python
# change color palette

sb.catplot(x = "size", y = "tip", data = df, hue = "sex",
          kind = "bar", palette = "Accent")
plt.show()
```

```python
# height and width

sb.catplot(x = "size", y = "tip", data = df, hue = "sex",
          kind = "bar", palette = "Accent", height = 10)
plt.show()
```

```python
# different kind of plots

sb.catplot(x = "size", y = "tip", data = df, hue = "sex",
          kind = "point", palette = "Accent")
plt.show()
```

# Styling Plot in Seaborn

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
df = sb.load_dataset("tips")
```

```python
df.head()
```

```python
sb.boxplot(y = df["total_bill"])
plt.show()
```

```python
# add grid

sb.barplot(x = "day", y = "total_bill", data = df)
plt.grid()
plt.show()
```

```python
# add grid using seaborn

sb.set(style = "darkgrid")
sb.barplot(x = "day", y = "total_bill", data = df)

plt.show()
```

```python
# remove axis params

sb.set(style = "white")
sb.barplot(x = "day", y = "total_bill", data = df)
sb.despine()
plt.show()
```

```python
# fig size change

plt.figure(figsize = (12,3))
sb.set(style = "darkgrid")
sb.barplot(x = "day", y = "total_bill", data = df)

plt.show()
```

```python
# change contents

sb.set_context("paper", font_scale = 5)
sb.set(style = "darkgrid")
sb.barplot(x = "day", y = "total_bill", data = df)

plt.show()
```

```python
# change colors
# cant assign palette without adding hue

sb.set(style = "darkgrid")
sb.barplot(x = "day", y = "total_bill", data = df, hue = "day", palette = "cool")

plt.show()
```

# Facet Grid plot

```python
import matplotlib.pyplot as plt
import seaborn as sb
import pandas as pd
```

```python
df = sb.load_dataset("tips")
```

```python
df.head()
```

```python
fg = sb.FacetGrid(df)
fg.map(plt.scatter, "total_bill", "tip")
plt.show()
```

```python
# adding mutiple plots

fg = sb.FacetGrid(df, col = "sex")
fg.map(plt.scatter, "total_bill", "tip")
plt.show()
```

```python
fg = sb.FacetGrid(df, col = "sex", hue = "day")
fg.map(plt.scatter, "total_bill", "tip")
plt.show()
```

```python
fg = sb.FacetGrid(df, col = "sex", hue = "day")
fg.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sb.FacetGrid(df, col = "day", hue = "sex")
fg.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

```python
fg = sb.FacetGrid(df, col = "day", hue = "sex")
fg.map(plt.bar, "total_bill", "tip").add_legend()
plt.show()
```

```python
# changing color palette

fg = sb.FacetGrid(df, col = "day", hue = "sex", palette = "summer")
fg.map(plt.scatter, "total_bill", "tip").add_legend()
plt.show()
```

```python
# adding edge color

fg = sb.FacetGrid(df, col = "day", hue = "sex", palette = "summer")
fg.map(plt.scatter, "total_bill", "tip", edgecolor = "r").add_legend()
plt.show()
```
